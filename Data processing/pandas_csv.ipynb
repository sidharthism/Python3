```python
df = pd.read_csv('homes.csv')
df.head()
```

```python
columns = list(df)
columns
```

```python
for r in df.iterrows():
    print(r)
```

```python
plt.scatter(np.arange(len(df)),df['Age'])
plt.show()
```

##### Deleting columns

```python
filtered_df = df.drop(['Taxes', 'Acres', 'List'], axis = 1)
filtered_df.head()
```

##### Filtering data

```python
filtered_df = filtered_df[filtered_df['Sell'] < 120 ]
filtered_df
```

##### Add a new row

```python
filtered_df = filtered_df.append(pd.DataFrame([{'Sell': 98, 'Living': 15, 'Rooms': 5, 'Beds': 4, 'Baths': 2, 'Age': 25}]))
filtered_df
```

Using pd.concat

```python
new_row = pd.DataFrame([{'Sell': 98, 'Living': 15, 'Rooms': 5, 'Beds': 4, 'Baths': 2, 'Age': 25}]) # Can specify index = [0] or any other value
filtered_df = pd.concat([new_row, filtered_df])
filtered_df
```

```python
plt.plot(filtered_df.index, filtered_df['Sell'])
plt.show()
```

```python
filtered_df.index.name = 'SNo'
filtered_df.to_csv('filtered_homes.csv')
```
