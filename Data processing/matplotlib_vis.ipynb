#### Matplotlib is easy to use visualizing library in Python. It is built on NumPy arrays and designed to work with the broader SciPy stack and consists of several plots like line, bar, scatter, histogram, etc. 

##### Basic plot

```python
# %pip install numpy
# %pip install matplotlib

import matplotlib.pyplot as plt
import numpy as np
 
# initializing the data
x = np.arange(10, 50, 10) # [10, 20, 30, 40]
y = x + 10 # [20, 30, 40, 50]
 
#Adding the title
plt.title("Simple plot")

# plotting the data
plt.plot(x, y)

plt.show()
```

##### Labels

```python
#Adding the title
plt.title("Simple plot with labels")

# plotting the data
plt.plot(x, y)

# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

##### Legends
A legend is an area describing the elements of the graph. In simple terms, it reflects the data displayed in the graph’s Y-axis. It generally appears as the box containing a small sample of each color on the graph and a small description of what this data means.

```python
#Adding the title
plt.title("Simple plot with labels and legends")

# plotting the data
plt.plot(x, y)
plt.plot(y, x)

# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")

plt.legend([" y = f(x)", "x = f(y)"]) # Should be given as an array

plt.show()
```

##### Ticks
Ticks are the values used to show specific points on the coordinate axis. It can be a number or a string. Whenever we plot a graph, the axes adjust and take the default ticks.

```python
#Adding the title
plt.title("Simple plot with labels, legends and ticks")

# plotting the data
plt.plot(x, y)
plt.plot(y, x)

# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")

plt.legend([" y = f(x)", "x = f(y)"]) # Should be given as an array

plt.xticks(np.arange(0, 100, 10))
plt.yticks(np.arange(0, 100, 10))

plt.show()
```

##### Bar graph

```python
#Adding the title
plt.title("Bar graph")

# plotting the data
plt.bar(x, y)

# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")

plt.xticks(np.arange(0, 50, 10))

plt.show()
```

##### Scatter plot

```python
#Adding the title
plt.title("Scatter plot")

# plotting the data
plt.scatter(x, y)

# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")

plt.xticks(np.arange(0, 50, 10))

plt.show()
```

##### Pie chart

```python
#Adding the title
plt.title("Pie chart")

# plotting the data
plt.pie(y)

plt.show()
```
